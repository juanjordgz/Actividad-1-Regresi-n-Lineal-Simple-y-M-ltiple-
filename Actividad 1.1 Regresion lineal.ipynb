```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Datos_Rio_Limpio.csv")
df.head(5)
```

df general

```python
cuantitativas = df.select_dtypes(include=["int64", "float64", "int32", "float32", "int", "float"])
print("Columnas cuantitativas:\n", cuantitativas.columns.tolist())
```

```python
pd.options.display.max_columns = None
```

```python
Corr_Factorsdf=cuantitativas.corr()
Corr_Factorsdf
```

```python
Corr_factordf= abs(Corr_Factorsdf)
Corr_factordf
```

```python
var_dep = "review_scores_rating"


correls = df.corr(numeric_only=True)[var_dep].drop(var_dep)  


correls_abs = correls.abs().sort_values(ascending=False)


print(f"Variables mÃ¡s correlacionadas con {var_dep}:\n")
print(correls_abs.head(5))


print("\nCon signo original:\n")
print(correls.loc[correls_abs.index[:5]])
```

```python
df1 = df[df['room_type'] == 'Entire home/apt']
df2 = df[df['room_type'] == 'Private room']
df3 = df[df['room_type'] == 'Shared room']
df4 = df[df['room_type'] == 'Hotel room']
```

Entire Home/apt df1

```python
cuantitativasdf1 = df1.select_dtypes(include=["int64", "float64", "int32", "float32", "int", "float"])
print("Columnas cuantitativas:\n", cuantitativas.columns.tolist())
```

```python
Corr_Factorsdf1=cuantitativasdf1.corr()
Corr_Factorsdf1
```

```python
pd.options.display.max_columns = None
```

```python
Corr_factordf1= abs(Corr_Factorsdf1)
Corr_factordf1
```

```python
pares = [
    ("host_acceptance_rate", "host_response_rate"),
    ("review_scores_rating", "calculated_host_listings_count"),
    ("host_acceptance_rate", "price"),
    ("availability_365", "number_of_reviews"),
    ("host_acceptance_rate", "number_of_reviews"),
    ("reviews_per_month", "review_scores_communication"),
]

for x, y in pares:
    if x in df1.columns and y in df1.columns:
        corr = df1[[x, y]].corr().iloc[0,1]
        print(f"CorrelaciÃ³n entre {x} y {y}: {abs(corr):.6f}")
```

```python

num_df1 = df1.select_dtypes(include=["int64","float64","int32","float32","int","float"])
num_df1 = num_df1.loc[:, ~num_df1.columns.str.contains('^Unnamed')]

corr_df1 = num_df1.corr().abs()

mask = np.triu(np.ones(corr_df1.shape), k=1).astype(bool)
corr_triu_df1 = corr_df1.where(mask)

corr_pairs_df1 = corr_triu_df1.stack().reset_index()
corr_pairs_df1.columns = ["Variable 1", "Variable 2", "CorrelaciÃ³n_abs"]

top10_df1 = corr_pairs_df1.sort_values(by="CorrelaciÃ³n_abs", ascending=False).head(10)

print("\nTop 10 correlaciones en df1 (Entire home/apt):\n")
display(top10_df1)

```
